# Data Preview

## Setup

```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "../data/Fitabase Data 4.12.16-5.12.16/"
```

## Tables

Id     | Table                    | Description
--     | ------------------------ | -----------
**1**  | dailyActivity            | Contains daily steps, activity distance/time, and calories.
**2**  | dailyCalories            | Daily calories burned. Included in **1**.
**3**  | dailyIntensities         | Daily breakdown of minutes/distance per activity intensity. Included in **1**.
**4**  | dailySteps               | Daily steps taken. Included in **1**.
**5**  | heartrate_seconds        | Heart rate for every 5 seconds.
**6**  | hourlyCalories           | Hourly calories burned.
**7**  | hourlyIntensities        | Hourly total and average intensity.
**8**  | hourlySteps              | Hourly step total.
**9**  | minuteCaloriesNarrow     | Hourly calories burned in long form.
**10** | minuteCaloriesWide       | Hourly calories burned in wide form (column per minute).
**11** | minuteIntensitiesNarrow  | Intensities per minute in long form.
**12** | minuteIntensitiesWide    | Intensities per minute in wide form (column per minute).
**13** | minuteMETsNarrow         | Metabolic Equivalent of Task (oxygen consumption, per minute, 1 == person at rest).
**14** | minuteSleep              | Sleep by minute.
**15** | minuteStepsNarrow        | Steps by minute.
**16** | minuteStepsWide          | Steps by minute in wide form (column per minute).
**17** | sleepDay                 | Daily sleep values. Includes # of sleep records and total minutes in bed vs. asleep.
**18** | weightLogInfo            | Weight logs in kg & lbs, fat%, BMI, and whether manually entered.

```python
# Show the top of each table

filenames = os.listdir(path)

for i, file in enumerate(filenames):
    print(f"{i + 1} : {file}")

    df = pd.read_csv(path + file)

    display(df.head())

    print("\n" + "*" * 80 + "\n")
```
